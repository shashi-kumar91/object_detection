```python
# Import the YOLO class from the ultralytics package
from ultralytics import YOLO

# Load a YOLOv8 model (using the nano version for speed)
model = YOLO('yolov8n.pt')  # You can choose 'yolov8s.pt' for small, 'yolov8l.pt' for large, etc.

# Train the model
model.train(
    data='/content/drive/MyDrive/Dataset/data.yaml',  # Path to your data.yaml file
    epochs=50,                                        # Number of epochs to train for
    imgsz=640,                                       # Image size (you can adjust this based on your system)
    batch=16,                                        # Batch size (you can adjust this based on your GPU memory)
    save=True                                        # Save the model after training
)

# Optional: Evaluate the model after training
metrics = model.val()
```

```python
metrics = model.val()  # This will run validation on the dataset and output metrics like mAP, precision, and recall
```

```python
results = model.predict(source='/content/drive/MyDrive/Dataset/test/images/1004_jpg.rf.8993d602c9a9b9bde6e32cb27d99007d.jpg')
```

```python
import matplotlib.pyplot as plt

# Run the model on the image
results = model.predict(source='/content/drive/MyDrive/Dataset/test/images/1004_jpg.rf.8993d602c9a9b9bde6e32cb27d99007d.jpg')

# Plot the image with predictions
plt.imshow(results[0].plot())  # results[0] gives the first image result, and .plot() draws the boxes and labels
plt.axis('off')  # Hide axis
plt.show()  # Display the image
```
